```python
# Problem 3: Mean Squared Error

# Store squared errors
squared_errors = [4, 9, 16, 25, 36]

# Calculate sum of squared errors
total_squared_error = squared_errors[0] + squared_errors[1] + squared_errors[2] + squared_errors[3] + squared_errors[4]

# Calculate Mean Squared Error
MSE = total_squared_error / 5
print("Mean Squared Error (MSE) =", MSE)
```

```python
# Problem 4: Cook's Distance

# Given values
ei_squared = 9
MSE = 3
p = 3
hii = 0.2
n = 25

# Calculate Cook's Distance
cooks_distance = (ei_squared / (p * MSE)) * (hii / (1 - hii) ** 2)
print("Cook's Distance =", cooks_distance)

# Threshold
threshold = 4 / n
print("Threshold =", threshold)

# Decision
print("Is the observation influential?", cooks_distance > threshold)
```
